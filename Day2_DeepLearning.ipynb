```python
#15
arr1=np.array([[1,2,3],[4,5,6]])
arr2=np.array([[7,8,9],[10,11,12]])

sumarr=arr1+arr2
subarr=arr1-arr2
prodarr=arr1*arr2
tranarr1=np.transpose(arr1)
tranarr2=np.transpose(arr2)
print('sum:\n',sumarr)
print('sub:\n',subarr)
print('prod:\n',prodarr)
print('trans1:\n',tranarr1)
print('trans2:\n',tranarr2)
```

```python
width, height = image_resize.size


left = width * 0.25
top = height * 0.25
right = width * 0.75
bottom = height * 0.75

center_cropped_image = image_resize.crop((left, top, right, bottom))

center_cropped_image
```
